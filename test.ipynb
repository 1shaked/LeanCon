```python
# for each item we will get his guid , name , type , category , unit , length , area , volume
table = []
for element in model.by_type("IfcProduct"):
    guid = element.GlobalId
    name = element.Name
    type = element.ObjectType
    category = element.is_a()
    psets = ifcopenshell.util.element.get_psets(element)
    length = None
    area = None
    volume = None
    if "Pset_LengthCommon" in psets:
        length = psets["Pset_LengthCommon"].get("Length", None)
    if "Pset_AreaCommon" in psets:
        area = psets["Pset_AreaCommon"].get("Area", None)
    if "Pset_VolumeCommon" in psets:
        volume = psets["Pset_VolumeCommon"].get("Volume", None)
    table.append(dict(guid=guid, name=name, type=type, category=category, length=length, area=area, volume=volume))
```

```python
table = []
for element in model.by_type("IfcProduct"):
    guid     = element.GlobalId
    name     = element.Name
    obj_type = element.ObjectType
    category = element.is_a()

    # get property sets + quantity sets
    psets = ifcopenshell.util.element.get_psets(element)
    qtos  = ifcopenshell.util.element.get_psets(element, qtos_only=True)

    length = None
    area   = None
    volume = None

    # existing logic for property sets
    if "Pset_LengthCommon" in psets:
        length = psets["Pset_LengthCommon"].get("Length", None)
    if "Pset_AreaCommon" in psets:
        area = psets["Pset_AreaCommon"].get("Area", None)
    if "Pset_VolumeCommon" in psets:
        volume = psets["Pset_VolumeCommon"].get("Volume", None)

    # **new logic**: check in quantity-sets (qtos)
    # example: a quantity set named "Qto_ColumnBaseQuantities"
    if "Qto_ColumnBaseQuantities" in qtos:
        q = qtos["Qto_ColumnBaseQuantities"]
        # q is a dict of quantity name → value
        # check for CrossSectionArea, GrossVolume, Length, OuterSurfaceArea
        length_q = q.get("Length", None)
        if length_q is not None:
            length = length_q
        area_q = q.get("CrossSectionArea", None)
        if area_q is not None:
            area = area_q
        volume_q = q.get("GrossVolume", None)
        if volume_q is not None:
            volume = volume_q

    # append row with everything
    table.append({
        "guid":     guid,
        "name":     name,
        "type":     obj_type,
        "category": category,
        "length":   length,
        "area":     area,
        "volume":   volume,
        # you could also store all quantities/psets if you like
        # "qtos":     qtos,
        # "psets":    psets,
    })
```

```python
data = {}

def get_level(element):
    """Return level name from spatial containment"""
    try:
        for rel in element.ContainedInStructure:
            storey = rel.RelatingStructure
            if storey and storey.is_a("IfcBuildingStorey"):
                return storey.Name
    except:
        pass
    return "No Level"

def extract_quantity(element):
    """Return (quantity_value, unit_string) or (None, None)"""
    qtos = ifcopenshell.util.element.get_psets(element, qtos_only=True)

    # The most common quantity set for structural elements:
    q = qtos.get("Qto_ColumnBaseQuantities", {}) \
        or qtos.get("Qto_BeamBaseQuantities", {}) \
        or qtos.get("Qto_WallBaseQuantities", {}) \
        or qtos.get("Qto_SlabBaseQuantities", {})

    # Priority: Volume → Area → Length
    if "GrossVolume" in q:
        return q["GrossVolume"], "m³"
    if "NetVolume" in q:
        return q["NetVolume"], "m³"
    if "CrossSectionArea" in q:
        return q["CrossSectionArea"], "m²"
    if "OuterSurfaceArea" in q:
        return q["OuterSurfaceArea"], "m²"
    if "Length" in q:
        return q["Length"], "m"
    
    return None, None

for element in model.by_type("IfcProduct"):
    # # We only want elements that represent buildable physical components
    # if not element.is_a(("IfcColumn", "IfcBeam", "IfcWall", "IfcSlab")):
    #     continue

    quantity, unit = extract_quantity(element)
    if quantity is None:
        continue

    # Determine type name (row identifier)
    element_type = element.ObjectType or element.Name or element.is_a()

    level = get_level(element)

    # Initialize if not exists
    if element_type not in data:
        data[element_type] = {
            "unit": unit,
            "project_total": 0.0,
            "levels": {}
        }

    # Update totals
    data[element_type]["project_total"] += quantity
    data[element_type]["levels"][level] = data[element_type]["levels"].get(level, 0.0) + quantity

# ---------------- DISPLAY RESULT TABLE ----------------
print("\n=== ELEMENT QUANTITY SUMMARY ===")
for etype, info in data.items():
    print(f"\nElement Type: {etype}")
    print(f"  Unit of Measure: {info['unit']}")
    print(f"  Total Amount in Project: {info['project_total']:.3f} {info['unit']}")
    print("  Per Level:")
    for lvl, val in info["levels"].items():
        print(f"    {lvl}: {val:.3f} {info['unit']}")
```

```python
target_type_name = "M_Concrete-Rectangular Beam:400 x 800mm"
target_level_name = "03 - Floor"  # or “Level 03” / “03 – Floor” depending on how the IFC storey is named

guids = []

def get_level(element):
    for rel in getattr(element, "ContainedInStructure", []):
        storey = rel.RelatingStructure
        if storey and storey.is_a("IfcBuildingStorey"):
            return storey.Name
    return None

for element in model.by_type("IfcProduct"):
    if element.ObjectType == target_type_name or element.Name == target_type_name:
        level = get_level(element)
        if level == target_level_name:
            guids.append(element.GlobalId)

print("GUIDs of", target_type_name, "in level", target_level_name, ":")
for g in guids:
    print("  ", g)
```
